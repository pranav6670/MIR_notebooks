## What is MIR?
* MIR is concerned with extraction and analysis of any kind of music entity(a song, instrumental cover, etc) on any representation level(an audio source(.wav, .mp3 formats), MIDI representation, etc).
* MIR is an interdisciplinary science aimed to studying the processes, systems and knowledge representations required for retrieving information from music.
* MIR research takes advantage of technologies and knowledge derived from signal processing, machine learning, music cognition, database management, human-computer interaction, music archiving or sociology of music.

## Computational factors in Music Perception
1. **Music Content** includes all the low-level content which can be extracted(e.g. temporal data, spectral data, loudness, etc).
2. **Music Context** defines additional metadata that cannot be extracted from the audio signal(e.g. lyrics, artists, etc).
3. **Listener** state includes the user state in a given moment(e.g.  mood, musical knowledge, etc).
4. **Listener Context** refers to the environment where the listener is in at a given moment(e.g. geography, society, etc).
